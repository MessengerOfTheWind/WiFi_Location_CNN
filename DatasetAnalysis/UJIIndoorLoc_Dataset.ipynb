# UJIIndoorLoc Analysis
---

https://archive.ics.uci.edu/dataset/310/ujiindoorloc

The UJIIndoorLoc dataset is a general database to compare indoor localization methodologies based on WiFi fingerprinting.  

The database covers three Universitat Jaume I buildings, each containing four or more floors covering roughly 110,000m2. This dataset can be used for classification, such as building and floor identification, or regression, where you estimate longitude and latitude.

From the initial look at the data and the additional variable information found on the dataset archive, we can better understand each attribute in the dataset. The dataset contains 529 columns and 19937 rows, the columns are made up of 520 WAP attributes, longitude, latitude, floor, buildingID, spaceID, relative position, userID, phoneID and a timestamp. All of these columns contain data of type int64.  

Each **WAP** (wireless access point) attribute contains a negative integer value from -104 to 0 or can be +100, the +100 value is used when a wireless access point is undetected. The archive states that *"Each WiFi fingerprint can be characterized by the detected Wireless Access Points (WAPs) and the corresponding Received Signal Strength Intensity (RSSI). The intensity values are represented as negative integer values ranging -104dBm (extremely poor signal) to 0dbM"*.

The **Longitude** attribute contains negative real values from -7695.9387549299299000 to -7299.786516730871000.  

The **Latitude** attribute contains positive real values from 4864745.7450159714 to 4865017.3646842018.  

The **Floor** attribute describes the altitude in floors inside the building, containing integer values ranging from 0 to 4.  

**BuildingID** is used to identify the building, measures were taken from three different buildings so it is a categorical integer value from 0 to 2.  

**SpaceID** identifies the space (office, corridor, classroom) where the capture was taken, they are also categorical integer values.  

**RelativePosition** describes the location with respect to the space (1 - Inside, 2 - Outside the space).  

**UserID** is a categorical integer value that describes which user is connected to a given entry in the database.  

**PhoneID** is also a categorical integer value describing the Android device used to record the data.  

**Timestamp** is the UNIX Time when a given capture was taken (Integer value).

# Exploratory Data Analysis

```python
# Identify index locations for each building in the dataframe, allows data to be plotted faster
build_0_index = train_df.index[train_df['BUILDINGID'] == 0].tolist()
build_1_index = train_df.index[train_df['BUILDINGID'] == 1].tolist()
build_2_index = train_df.index[train_df['BUILDINGID'] == 2].tolist()
```

```python
def make_plot(data,plot_info,figsize,name):
    b0_component1 = [data[x,0] for x in build_0_index]
    b0_component2 = [data[x,1] for x in build_0_index]

    b1_component1 = [data[x,0] for x in build_1_index]
    b1_component2 = [data[x,1] for x in build_1_index]

    b2_component1 = [data[x,0] for x in build_2_index]
    b2_component2 = [data[x,1] for x in build_2_index]
    
    fig, ax = plt.subplots(figsize=figsize)
    colors = ['tab:blue','tab:orange','tab:green']
    marks = ['^','o','s']

    ax.scatter(b0_component1, b0_component2,color=colors[0],marker=marks[0],edgecolor='black')
    ax.scatter(b1_component1, b1_component2,color=colors[1],marker=marks[1],edgecolor='black')
    ax.scatter(b2_component1, b2_component2,color=colors[2],marker=marks[2],edgecolor='black')

    ax.set_xlabel(plot_info[0]) 
    ax.set_ylabel(plot_info[1])
    ax.set_title(plot_info[2])
    
    plt.legend(['Building 0','Building 1','Building 2'])
    plt.savefig('dataset_analysis_plots/'+name)
    plt.show()
```

## Principal Component Analysis

```python
#Experiment
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pcaUJI = PCA(n_components=2)
sc = StandardScaler()
```

```python
reduced_train = train_df[train_df.columns[520:]] # Excludes 0-520 WAP Columns 
scaled_train = sc.fit_transform(reduced_train)
pcaUJI.fit(scaled_train)
scores = pcaUJI.transform(scaled_train)
```
