```python
# 统计参考点个数
import pandas as pd

def count_reference_points(df, building_col='BUILDINGID', lon_col='LONGITUDE', lat_col='LATITUDE'):
    # 保留经纬度组合唯一
    df_unique = df[[building_col, lon_col, lat_col]].drop_duplicates()

    # 统计每栋楼参考点数量
    result = (df_unique
              .groupby(building_col)
              .size()
              .reset_index(name='num_reference_points'))

    # 排序（按建筑编号升序）
    result = result.sort_values(by=building_col).reset_index(drop=True)

    return result
```

```python
# 分别统计 Train / Valid / Test
train_rp = count_reference_points(train)
valid_rp = count_reference_points(validat)
# test_rp  = count_reference_points(test_df)

print("Train RP:\n", train_rp)
print("Valid RP:\n", valid_rp)
# print("Test  RP:\n", test_rp)
```

```python
# 统计楼层
import pandas as pd

def count_floors_per_building(df, building_col='BUILDINGID', floor_col='FLOOR'):
    result = (
        df[[building_col, floor_col]]
        .drop_duplicates()                      # 同一楼层仅计一次
        .groupby(building_col)
        .size()
        .reset_index(name='num_floors')
        .sort_values(by=building_col)
        .reset_index(drop=True)
    )
    return result
```

```python
train_floor_stats = count_floors_per_building(train)
valid_floor_stats = count_floors_per_building(validat)
# test_floor_stats  = count_floors_per_building(test_df)

print(train_floor_stats)
print(valid_floor_stats)
# print(test_floor_stats)
```

```python
# 统计在各建筑的样本数
import pandas as pd

def count_samples_per_building(df, building_col='BUILDINGID'):
    result = (
        df[[building_col]]
        .groupby(building_col)
        .size()
        .reset_index(name='num_samples')
        .sort_values(by=building_col)
        .reset_index(drop=True)
    )
    return result
```

```python
train_stats = count_samples_per_building(train)
valid_stats = count_samples_per_building(validat)
# test_stats  = count_samples_per_building(test_df)

print("Train:\n", train_stats)
print("Valid:\n", valid_stats)
# print("Test:\n", test_stats)
```
